```python
#import modules
import os
import glob
import pandas as pd

path = "CSV Data files/00_All CSV/20-21"

#list all csv files only
csv_files = glob.glob(path + "/*.csv")

```

```python
for CSV_File in csv_files: 
    
    # Dynamically create dataframes for each sheet in the excel file.
    vars()[CSV_File] = pd.DataFrame()

    # Read the excel sheet into a Pandas dataframe
    vars()[CSV_File] = pd.read_csv(CSV_File)
    display(vars()[CSV_File].head(2))
```

```python
df_concat = pd.concat([pd.read_csv(f) for f in csv_files ], ignore_index=False)
df_concat.info()

```

```python
df_concat.drop("Unnamed: 0", axis = 1)
```

```python
df_concat.to_csv('CSV Data files/2020-2021_CBS.csv')
```
